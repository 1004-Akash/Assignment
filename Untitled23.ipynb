```python
import matplotlib.pyplot as plt
import numpy as np
# Example data
data = [50, 55, 60, 65, 70, 75, 80, 5]
# Calculate Q1, Q3, and IQR
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
print(Q1)
print(Q3)
IQR = Q3 - Q1
print(IQR)
# Calculate bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Plotting boxplot
plt.figure(figsize=(8, 6))
plt.boxplot(data, vert=False, patch_artist=True,
            boxprops=dict(facecolor="lightblue", color="blue"),
            medianprops=dict(color="red"),
            whiskerprops=dict(color="blue"),
            capprops=dict(color="blue"),
            flierprops=dict(marker="o", color="red", markersize=8))

# Adding annotations for bounds
plt.axvline(lower_bound, color='orange', linestyle='--', label=f"Lower Bound ({lower_bound:.2f})")
plt.axvline(upper_bound, color='green', linestyle='--', label=f"Upper Bound ({upper_bound:.2f})")

# Adding labels and legend
plt.title("Boxplot with Lower and Upper Bound Annotations", fontsize=14)
plt.xlabel("Values", fontsize=12)
plt.legend(fontsize=10)
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.tight_layout()

# Show plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Example DataFrame
data = {
    "Feature1": [10, 12, 14, 15, 18, 20, 30, 35, 100, 50],
    "Feature2": [5, 7, 9, np.nan, 12, 15, 80, 60, 90, 8]
}
df = pd.DataFrame(data)

print("Original DataFrame:\n", df)

# Boxplot to visualize outliers
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, orient="h")
plt.title("Boxplot of Features")
plt.show()

# Identify outliers using IQR
def identify_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

outliers_feature1 = identify_outliers(df, "Feature1")
outliers_feature2 = identify_outliers(df, "Feature2")
print("\nOutliers in Feature1:\n", outliers_feature1)
print("\nOutliers in Feature2:\n", outliers_feature2)

# Remove outliers
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

df_cleaned = remove_outliers(df, "Feature1")
df_cleaned = remove_outliers(df_cleaned, "Feature2")
print("\nDataFrame after removing outliers:\n", df_cleaned)

# Fill missing values with standard values (e.g., mean or median)
df_cleaned["Feature2"].fillna(df_cleaned["Feature2"].mean(), inplace=True)
print("\nDataFrame after filling missing values:\n", df_cleaned)

# Cap values at 5th and 95th percentiles
def cap_values(df, column):
    lower_cap = df[column].quantile(0.05)
    upper_cap = df[column].quantile(0.95)
    df[column] = np.clip(df[column], lower_cap, upper_cap)
    return df

df_capped = cap_values(df_cleaned, "Feature1")
df_capped = cap_values(df_capped, "Feature2")
print("\nDataFrame after capping values:\n", df_capped)

# Boxplot after preprocessing
plt.figure(figsize=(10, 5))
sns.boxplot(data=df_capped, orient="h")
plt.title("Boxplot After Preprocessing")
plt.show()


```
